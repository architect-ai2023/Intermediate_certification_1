1.Какой регион потребляет максимальную и минимальную мощность в год?

```python
c=dict()
for i in (long.Regions.unique()):
    a=long[(long["Regions"]==i)&(long["year"]==2020)]
    b=a.Usage.sum()
    c.update({i:b})
    
print(c)

d=dict()
for i in (long.Regions.unique()):
    a=long[(long["Regions"]==i)&(long["year"]==2019)]
    b=a.Usage.sum()
    d.update({i:b})
    
print(d)
```

```python
print(f" максимальная потребляемая мощность в 2020 is {c[max(c)]} by {max(c)} регион\n минимальная потребляемая мощность в 2020 is {c[min(c)]} by {min(c)} регион  ")
print(f" максимальная потребляемая мощность в 2019 is {d[max(d)]} by {max(d)} регион\n минимальная потребляемая мощность в 2019 is {d[min(d)]} by {min(d)} регион  ")
```

```python
table1=pd.DataFrame.from_dict(c,orient="index")
table2=pd.DataFrame.from_dict(d,orient="index")
```

```python
table1.rename(columns={0:"total usages (2020)"},inplace=True)
table2.rename(columns={0:"total usages (2019)"},inplace=True)
```

```python
table1.head(2)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="total usages (2020)",y=table1.index,data=table1)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="total usages (2019)",y=table2.index,data=table2)
```

Каково самое высокое значение, достигнутое каждым штатом в любое время года в 2019 году?

```python
data.head(2)
```

```python
table10=data[data["year"]!=2020]
```

```python
states=list(table10.columns)
states.remove('year')
states.remove('month')
states.remove('day')
```

```python
for i in states:
    x=max(table10[i])
    print(f"наивысшее значение, достигнутое штатом {i} \t в \t 2019 году \t {x}")
```

```python
dict3=dict()
for i in states:
    x=max(table10[i])
    dict3.update({i:x})
    
table11=pd.DataFrame.from_dict(dict3,orient="index")
table11.rename(columns={0:"достигнуто наивысшее значение"},inplace=True)
table11.head(3)
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x="достигнуто наивысшее значение",y=table11.index,data=table11)
```

Каково самое низкое значение, достигнутое каждым штатом в любое время года в 2019 году?

```python
for i in states:
    x=min(table10[i])
    print(f"наименьшее значение, достигнутое штатом {i} \t в \t 2019 году \t {x}")
```

```python
dict4=dict()
for i in states:
    x=min(table10[i])
    dict4.update({i:x})
    
table12=pd.DataFrame.from_dict(dict4,orient="index")
table12.rename(columns={0:"наименьшее значение"},inplace=True)
table12.head(3)
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x="наименьшее значение",y=table12.index,data=table12)
```
